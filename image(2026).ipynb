```python
import cv2
import matplotlib.pyplot as plt

from google.colab.patches import cv2_imshow
import cv2
from google.colab import files
```

```python
# Upload an image file
uploaded = files.upload()
```

```python
# Get the file name of the uploaded image
image_path = list(uploaded.keys())[0]
```

```python
# Reading the image
image = cv2.imread(image_path)
```

```python
# Check if the image was loaded successfully
if image is None:
    print("Error: Could not load the image. Ensure the file path is correct.")
else:
    # Display the image
    cv2_imshow(image)
```

```python
# Writing the image to a new file
output_path ='rose1.jpg'
cv2.imwrite(output_path, image)
print(f"Image saved as {output_path}")
```

```python
# Read image using OpenCV
image = cv2.imread(image_path)  # Read in BGR format
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert to RGB

# Display the original image
plt.imshow(image_rgb)
plt.title("Original Image")
plt.axis("off")
plt.show()
```

```python
# Convert to grayscale
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Display both images
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
ax[0].imshow(image_rgb)
ax[0].set_title("RGB Image")
ax[0].axis("off")

ax[1].imshow(image_gray, cmap="gray")
ax[1].set_title("Grayscale Image")
ax[1].axis("off")

plt.show()

# Print pixel values
print("Pixel Value at (50,50) in RGB:", image_rgb[50, 50])
print("Pixel Value at (50,50) in Grayscale:", image_gray[50, 50])
```

```python
# Resize the image (scale down by 50%)
scaled_image = cv2.resize(image_rgb, None, fx=0.5, fy=0.5)

# Display
plt.imshow(scaled_image)
plt.title("Scaled Image (50%)")
plt.axis("off")
plt.show()
```

```python
# Get image center and define rotation matrix
(h, w) = image_rgb.shape[:2]
center = (w // 2, h // 2)
rotation_matrix = cv2.getRotationMatrix2D(center, 45, 1)  # Rotate by 45 degrees

# Apply rotation
rotated_image = cv2.warpAffine(image_rgb, rotation_matrix, (w, h))

# Display rotated image
plt.imshow(rotated_image)
plt.title("Rotated Image (45)")
plt.axis("off")
plt.show()
```

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt

# Read the image from image_path to ensure it's available in this cell's scope.
# The image_path variable holds the path to the uploaded image (e.g., '297657.jpg').
image = cv2.imread(image_path)

# Check if the image was loaded successfully before proceeding
if image is None:
    print("Error: Could not load the image from specified path.")
else:
    # Convert the image from BGR to HSV color space
    image_hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

    # Set the kernel size (can be adjusted)
    figure_size = 9

    # Apply a Gaussian blur with the specified kernel size
    new_image = cv2.GaussianBlur(image_hsv, (figure_size, figure_size), 0)

    # Display the original image
    plt.figure(figsize=(11,6))
    plt.subplot(121), plt.imshow(cv2.cvtColor(image_hsv, cv2.COLOR_HSV2RGB)), plt.title('Original')
    plt.xticks([]), plt.yticks([])

    # Display the image after Gaussian blur
    plt.subplot(122), plt.imshow(cv2.cvtColor(new_image, cv2.COLOR_HSV2RGB)), plt.title('Gaussian Blur')
    plt.xticks([]), plt.yticks([])

    plt.show()
```
